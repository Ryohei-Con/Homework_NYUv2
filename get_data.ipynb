```python
import zipfile
import os
from pathlib import Path
import urllib.request
import numpy as np
```

```python
# 例: 一部のパーツをリスト化してループ回す、または公式のインデックスファイルを読み込む
base_url = "http://horatio.cs.nyu.edu/mit/silberman/nyu_depth_v2/"
parts = ["basements.zip"]

for part in parts:
    print(f"Downloading {part}...")
    zip_path = Path(part + ".zip")
    urllib.request.urlretrieve(base_url + part, zip_path)
    extract_path = zip_path.stem
    print(F"解凍中: {part}")
    with zipfile.ZipFile(zip_path, "r") as zip_ref:
        zip_ref.extractcall(extract_path)
    print(f"解凍完了: {part}")
    os.remove(zip_path)
```
